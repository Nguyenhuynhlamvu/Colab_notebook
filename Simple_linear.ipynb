```python
data.plot(kind='scatter', x='wt',y='mpg',figsize=(9,9), color='black')
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
x=pd.DataFrame(data['wt'])
y=data['mpg']
model.fit(x,y)
print(model.intercept_)
print(model.coef_)
```

```python
model.score(x,y)
```

```python
y_pred=model.predict(x)
```

```python
data.plot(kind='scatter', x='wt',y='mpg',figsize=(9,9), color='black')
plt.plot(data['wt'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
x1=pd.DataFrame([data['wt'],data['wt']**3]).T

model.fit(x1,y=data['mpg'])
print("Intercept of model ", model.intercept_)
print('Coefficient of model', model.coef_)
print("Predict off model", model.score(x1,data['mpg']))
```

```python
range=np.arange(1.5,5.6,0.1)
predictor=pd.DataFrame([range,range**2]).T
y_pred=model.predict(predictor)
data.plot(kind="scatter",x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')
print(y_pred)
```
