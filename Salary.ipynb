```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data=pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/Salary_Data.csv')
```

```python
x=data.drop(['Salary'],axis=True)
y=data['Salary']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,x_test.shape, y_train.shape, y_test.shape
```

```python
sns.pairplot(data)
```

```python
plt.plot(x_train,y_train,'r',label='Train data')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend
plt.show()
```

```python
model=linear_model.LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test).round(1)
calculation=pd.DataFrame(np.c_[y_test,y_pred],columns=['original salary','predict salary'])
print(calculation)
```

```python
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,model.predict(x_train),color='blue')
plt.title("salary vs exxperience (train set)")
plt.xlabel('year vs experience')
plt.ylabel('salary')
plt.show()

print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
```
